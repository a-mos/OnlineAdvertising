```python
import pandas as pd
import torch
import numpy as np
from sklearn.metrics import log_loss, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
from deepctr_torch.inputs import SparseFeat, DenseFeat, VarLenSparseFeat, get_feature_names
from deepctr_torch.models import *
```

```python
def split_cg1(x):
    try:
        key_ans = x.split(',')
        for key in key_ans:
            if key not in key2index_cg1:
                # Notice : input value 0 is a special "padding",so we do not use 0 to encode valid feature for sequence input
                key2index_cg1[key] = len(key2index_cg1) + 1
        return list(map(lambda x: key2index_cg1[x], key_ans))
    except:
        return []
```

```python
def split_cg2(x):
    try:
        key_ans = x.split(',')
        for key in key_ans:
            if key not in key2index_cg2:
                # Notice : input value 0 is a special "padding",so we do not use 0 to encode valid feature for sequence input
                key2index_cg2[key] = len(key2index_cg2) + 1
        return list(map(lambda x: key2index_cg2[x], key_ans))
    except:
        return []
```

```python
def split_cg3(x):
    try:
        key_ans = x.split(',')
        for key in key_ans:
            if key not in key2index_cg3:
                # Notice : input value 0 is a special "padding",so we do not use 0 to encode valid feature for sequence input
                key2index_cg3[key] = len(key2index_cg3) + 1
        return list(map(lambda x: key2index_cg3[x], key_ans))
    except:
        return []
```

```python
data = dff

sparse_features = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10', 'C11', 'C12']
dense_features = ['l1', 'l2', 'hour', 'minutes']

data[sparse_features] = data[sparse_features].fillna('-1', )
data[dense_features] = data[dense_features].fillna(0, )
target = ['label']

# 1.Label Encoding for sparse features,and do simple Transformation for dense features
for feat in sparse_features:
    lbe = LabelEncoder()
    data[feat] = lbe.fit_transform(data[feat])
mms = MinMaxScaler(feature_range=(0, 1))
data[dense_features] = mms.fit_transform(data[dense_features])
```

```python
import tqdm
```

```python
key2index_cg1 = {}
cg1_list = list(map(split_cg1, tqdm.tqdm(data['CG1'].values)))
data = data.drop(['CG1'], axis=1)
data = data.drop(['CG2'], axis=1)
data = data.drop(['CG3'], axis=1)
cg1_length = np.array(list(map(len, cg1_list)))
cg1_max_len = max(cg1_length)
# Notice : padding=`post`
cg1_list = pad_sequences(cg1_list, maxlen=cg1_max_len, padding='post', )

# key2index_cg2 = {}
# cg2_list = list(map(split_cg2, tqdm.tqdm(data['CG2'].values)))
# data = data.drop(['CG2'], axis=1)
# cg2_length = np.array(list(map(len, cg2_list)))
# cg2_max_len = max(cg2_length)
# # Notice : padding=`post`
# cg2_list = pad_sequences(cg2_list, maxlen=cg2_max_len, padding='post', )

# key2index_cg3 = {}
# cg3_list = list(map(split_cg3, tqdm.tqdm(data['CG3'].values)))
# data = data.drop(['CG3'], axis=1)
# cg3_length = np.array(list(map(len, cg3_list)))
# cg3_max_len = max(cg3_length)
# # Notice : padding=`post`
# cg3_list = pad_sequences(cg3_list, maxlen=cg3_max_len, padding='post', )
```

```python
# 2.count #unique features for each sparse field,and record dense feature field name

fixlen_feature_columns = [SparseFeat(feat, data[feat].nunique(), embedding_dim=10)
                          for feat in sparse_features] + [DenseFeat(feat, 1) for feat in dense_features]

varlen_feature_columns = [VarLenSparseFeat(SparseFeat('CG1', vocabulary_size=len(key2index_cg1) + 1, embedding_dim=10), maxlen=cg1_max_len, combiner='mean')]  # Notice : value 0 is for padding for sequence input feature
#varlen_feature_columns += [VarLenSparseFeat(SparseFeat('CG2', vocabulary_size=len(key2index_cg2) + 1, embedding_dim=10), maxlen=cg2_max_len, combiner='mean')]
#varlen_feature_columns += [VarLenSparseFeat(SparseFeat('CG3', vocabulary_size=len(key2index_cg3) + 1, embedding_dim=10), maxlen=cg3_max_len, combiner='mean')]

dnn_feature_columns = fixlen_feature_columns + varlen_feature_columns
linear_feature_columns = fixlen_feature_columns + varlen_feature_columns

feature_names = get_feature_names(linear_feature_columns + dnn_feature_columns)
```

```python
# 3.generate input data for model
num_train = 29989752
X_train = data.iloc[:num_train]
y_train = data.iloc[:num_train]['label']

X_test = data.iloc[num_train:]
y_test = data.iloc[num_train:]['label']

train_model_input = {name: X_train[name] for name in feature_names if 'CG' not in name}
train_model_input["CG1"] = cg1_list[:num_train]
# train_model_input["CG2"] = cg2_list[:num_train]
# train_model_input["CG3"] = cg3_list[:num_train]


test_final = {name: X_test[name] for name in feature_names if 'CG' not in name}
test_final["CG1"] = cg1_list[num_train:]
# test_final["CG2"] = cg2_list[num_train:]
# test_final["CG3"] = cg3_list[num_train:]
```
